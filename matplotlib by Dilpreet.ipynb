```python
x=np.array([2,5,3,9])
y=np.array([5,4,6,7])
font1={"family":"serif","color":"black","size":20}
font2={"family":"serif","color":"red","size":10}
plt.title("Salary vs Experience",fontdict=font1)
plt.xlabel("Salary",fontdict=font2)
plt.ylabel("Experience",fontdict=font2)
plt.plot(x,y)
```
