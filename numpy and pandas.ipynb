```python
# How to deal with null or Nan values
```

```python
import pandas as pd
import numpy as np
```

```python
d={
    "Ages":[15,50,42,np.nan,45,56],
    "Weight":[45,np.nan,58,85,56,89],
    "Height":[np.nan,78,np.nan,45,56,36],
    "DOB":[1997,2001,2007,1998,2003,np.nan]
}
```

```python
d
```

```python
pd.DataFrame(d)
```

```python
df=pd.DataFrame(d)
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df.notnull()
```

```python
df.notnull().sum()
```

```python
df.fillna(0)
```

```python
df.fillna(1)
```

```python
df.fillna(method="bfill")
```

```python
df.fillna(method="ffill")
```

```python
df.interpolate(method="linear",limit_direction="both")
```

```python
df.replace(np.nan,4)
```
