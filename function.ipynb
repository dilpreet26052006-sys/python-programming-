```python
def info(*name):
    print("hello",name)
info("zubair","dilpreet","yash","saman","aman")
```

```python
def info(*name):
    print("hello",name[4])
info("zubair","dilpreet","yash","saman","aman")
```

```python
def info(**name):
    print("Hello my name is:",name)
info(name="Zubair")
```

```python
def info(**name):
    print("Hello",name["sname"])
info(fname="zubair",sname="fayaz",tname="dilpreet")
```
