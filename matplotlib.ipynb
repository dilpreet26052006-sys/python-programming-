```python
# Create a 2*2 grid of subplots
fig,axis=plt.subplots(2,2)
# Plot on each subplots
axis[0,1].bar([1,2,3],[4,5,6])
axis[0,1].set_title('Plot 2')
plt.tight_layout()
plt.show()
```

```python
# Data for the polar plot
theta=np.linspace(0,2*np.pi,100)
r=np.sin(theta) # Example of periodic data
# Create a polar plot
plt.subplot(projection='polar')
plt.plot(theta,r)
plt.title('Polar Plot')
plt.show()
```

```python
# Plot data
x=[1,2,3,4,5]
y=[2,3,5,7,11]
plt.plot(x,y)
# Annotate a specific point
plt.annotate("Prime number",xy=(3,5),xytext=(4,6),
arrowprops=dict(facecolor='red',arrowstyle='->'))
plt.title('Annotated Plot')
plt.show()
```

```python
# Create random data
data=np.random.rand(10,10)
# Create a heatmap
plt.imshow(data,cmap='viridis',interpolation='nearest')
plt.colorbar() # Show color scale
plt.title('Heatmap')
plt.show()
```

```python
# Create random data
data=np.random.rand(10,10)
# Create a heatmap
plt.imshow(data,cmap='prism',interpolation='nearest')
plt.colorbar() # Show color scale
plt.title('Heatmap')
plt.show()
```

```python
labels=['A','B','C','D']
sizes=[15,30,45,10]
plt.pie(sizes,labels=labels,autopct='%1.1f%%')
plt.title("Basic Pie Chart")
plt.show()
```
