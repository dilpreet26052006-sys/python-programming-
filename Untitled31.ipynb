```python
# factorial program
def fact(n):
    if n==0:
        return 1
    elif n==1:
        return 1
    else:
        return n* fact(n-1)
n=int(input("Enter the value of n:=")) 
print(fact(n))
```

```python
fact(5)
```

```python
def fib(n):
    if n==0:
        return 0
    elif n==1:
        return 1
    else:
        return fib(n-1)+(n-2)
n=int(input("Enter any number"))
for i in range(n):
    print(fib(i))
```

```python
fib(9)
```
