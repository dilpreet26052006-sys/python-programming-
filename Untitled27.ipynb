```python
n=int(input("enter the no"))
i=1
count=0
while i<=n:
    if n%i==0:
        count=count+1
        i=i+1
    else:
        i=i+1
else:
    if count==2:
        print("the no is a prime no")
    else:
        print("the given no is not a prime no")
```
